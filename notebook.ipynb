# ChatCompletion: Managing Roles with messages

## Helpful Chatbot:

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Who won the world series in 2002?"},
]
```

```python
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages = messages
)
print(response.choices[0]['message']['content'])
```

## Helpful Chatbot that speaks in shakespearean:

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant that speaks in Shakespearian language"},
    {"role": "user", "content": "Who won the world series in 2002?"},
]
```

```python
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages = messages
)
print(response.choices[0]['message']['content'])
```

# Customized Zoo Chatbot

```python
messages = [
    {
      "role": "system",
      "content": (
        "You are an assistant for an imaginary zoo called Hypo2oo. \n"
        "You have detailed knowledge about various animals, their habitats, \n"
        "feeding times, and the zoo's daily activities schedule."
      )
    },
    {
        "role": "user",
        "content": "What is the feeding time for the lions?"
    }
]
```

```python
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages = messages
)
```

```python
print(response.choices[0]['message']['content'])
```

```python
messages.append({
    "role": "assistant",
    "content": response.choices[0]['message']['content']
})
```

```python
messages.append({
    "role": "user",
    "content": "what about the elephants?"
})
```

```python
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages = messages
)
```

```python
print(response.choices[0]['message']['content'])
```

# Manipulating Temperature



## High Temperature

```python
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages = messages,
    temperature = 1.6
)
```

```python
print(response.choices[0]['message']['content'])
```

## Low Temperature

```python
response = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    messages = messages,
    temperature = 0
)
```

```python
print(response.choices[0]['message']['content'])
```

# Museum Visitor Chatbot

```python
messages = [
{
  "role": "system", "content": "You are an assistant at a museum that specializes in natural history \n"
    "and science exhibits. \n"
    "The museum attracts visitors of all ages, from curious children to enthusiastic adults. \n"
    "Your primary objective is to enhance the visitor experience by providing information, \n"
    "answering questions, and engaging in interactive conversations about the exhibits"
}
]
```

```python
messages.append({
  "role": "user",
  "content": "Tell me about this civil war exhibit."
})
```

```python
try:
    response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",
      messages=messages,
      max_tokens=256,
      temperature=0.6
    )
    print(response["choices"][0]["message"]["content"])
except Exception as e:
    print(f"Received an error from the OpenAI API: {e}")
```
